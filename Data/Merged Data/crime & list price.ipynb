```python
import pandas as pd
```

```python
merged_df = pd.read_csv("./Incidents & Avg List Price by zip code.csv")
merged_df
```

```python
merged_df.head(20)
```

```python
# Plotting average list price
# We need to normalize the average list price data to fit in the same scale as crime incidents
max_price = merged_df['Average List Price'].max()
price_scale_factor = max(merged_df['Crime Incidents']) / max_price
normalized_prices = merged_df['Average List Price'] * price_scale_factor
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Recalculate the normalized average list price
max_price = merged_df['Average List Price'].max()
price_scale_factor = max(merged_df['Crime Incidents']) / max_price
normalized_prices = merged_df['Average List Price'] * price_scale_factor

# Create a clustered bar chart using matplotlib
x = np.arange(len(merged_df.index))
width = 0.35

fig, ax = plt.subplots(figsize=(14, 7))

# Plotting crime incidents
ax.bar(x - width/2, merged_df['Crime Incidents'], width, color='red', label='Crime Incidents')

# Plotting average list price
ax.bar(x + width/2, normalized_prices, width, color='blue', label='Average List Price (Normalized)')

ax.set_xlabel('ZIP Code')
ax.set_ylabel('Count / Normalized Price')
ax.set_title('Crime Incidents and Normalized Average List Price by ZIP Code')
ax.set_xticks(x)
ax.set_xticklabels(merged_df.index)
ax.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
