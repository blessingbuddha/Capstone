```python
#import pandas
import pandas as pd

#read the csv with headers
rating_df = pd.read_csv('statewideoverallaccountabilityratings2022.csv', header=None)

#use iloc to gets rows (and/or columns) at integer locations 
#replace the new line with a space
rating_df.iloc[0] = rating_df.iloc[0].str.replace('\n'," ")

#make the columns lock in as an integer
rating_df.columns = rating_df.iloc[0]

#drop the new lines and make the space value true
rating_df.drop(0,inplace=True)

#save the new csv file
rating_df.to_csv('statewideoverallaccountabilityratings2022.csv')
```
